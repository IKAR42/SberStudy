### 1. Сколько вакансий, которые вам нравятся?

```python
#Определяем 2 условия 
    1. nice1 - наименование вакансии, которая включает в себя формулировку 'data analyst'
    2. nice2 - описание вакансии, которое включает в себя формулировку 'дружная команда'
#Отфилтровываем датафрейм с использованием 2 указанных условий
```

```python
nice1 = vacs.vactitle.str.lower().str.contains('data analyst')
nice2 = vacs.vacdescription.str.lower().str.contains('дружная команда')
vacs[nice1 & nice2]
```
