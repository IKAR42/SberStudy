### Определяем фичи и таргет

```python
x = df_housing.drop('MEDV',axis=1)
y = df_housing['MEDV']
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=80)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_medv = LinearRegression()
model_medv.fit(X_train,y_train)
```

```python
pred_medv = model_medv.predict(X_test)
```

```python
pred_medv
```

```python
y_test
```

### Считаем метрики и оцениваем качество модели

```python
print('Ошибка MSE=',mean_squared_error(y_test,pred_medv, squared=True))
```

```python
print('Ошибка MAE =',mean_absolute_error(y_test,pred_medv))
```
